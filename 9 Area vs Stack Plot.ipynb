```python
from matplotlib import pyplot as plt
```

# Stack Plot

```python
# minimum 2 values required for plotting stack plot
x = [1,2,3,4,5]
area = [1,2,3,4,5] # y axis

plt.stackplot(x,area)
plt.show()
```

```python
# by changing the values of x and area stack plot changes
x = [7,2,1,4,6]
area = [2,5,7,2,6] # y axis

plt.stackplot(x,area)
plt.show()
```

```python
# Plotting multiple stack plots in a plot
x = [1,2,3,4,5]
area1 = [1,2,3,4,5] # y axis
area2 = [2,3,8,9,1]
area3 = [5,4,8,1,2]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
# giving this a label
# label must be passed in a list
# label parameter is used in the stackplot(x,y,z,label=list)
# when we are using label we must use legend

x = [1,2,3,4,5]
area1 = [1,2,3,4,5] # y axis
area2 = [2,3,8,9,1]
area3 = [5,4,8,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.legend()

plt.show()
```

```python
# we can change the colors also

x = [1,2,3,4,5]
area1 = [1,2,3,4,5] # y axis
area2 = [2,3,8,9,1]
area3 = [5,4,8,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l, colors=['r','g','m'])

plt.legend()

plt.show()
```

```python
# changing baseline
# by default is baseline='zero'

x = [1,2,3,4,5]
area1 = [1,2,3,4,5] # y axis
area2 = [2,3,8,9,1]
area3 = [5,4,8,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l, baseline='sym')

plt.legend()

plt.show()
```

```python
# Giving plot a title
x = [1,2,3,4,5]
area1 = [1,2,3,4,5] # y axis
area2 = [2,3,8,9,1]
area3 = [5,4,8,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.legend()
plt.title('Python')

plt.show()
```

```python
# Giving plot x axis and y axis
x = [1,2,3,4,5]
area1 = [1,2,3,4,4] # y axis
area2 = [2,6,8,9,1]
area3 = [5,4,7,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.legend()
plt.title('Stack Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# giving this plot a grid

x = [1,2,3,4,5]
area1 = [1,2,3,4,4] 
area2 = [2,6,8,9,1]
area3 = [5,4,7,1,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.legend()
plt.title('Stack Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()

plt.show()
```
