```python
import matplotlib.pyplot as plt
```

# Stem Plot

```python
# creating values for Stem Plot
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y)
plt.show()
```

```python
# in above example lines are solid 
# we can change them using the description
from IPython import display
display.Image('img12.png')
```

```python
# changing line type
# using linefmt='value'
plt.stem(x,y,linefmt='-.')
plt.show()
```

```python
# we can change the ink of dot
# using markerfmt='colordottype' (r+,g*,and soon) color and and dot type

plt.stem(x,y,linefmt='-.', markerfmt='g*')
plt.show()
```

```python
# in above diagram y axis value is starting from 0
# we can start it by any other number 
# using bottom=integer

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=1)
plt.show()
```

```python
# another example

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=5)
plt.show()
```

```python
# in above examples base line color is red
# we can change it any another color
# using basefmt='color'

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=1, basefmt='g')
plt.show()
```

```python
# we can use a label also 
# using label='string'
# must use plt.legend() if using label

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=1, label='Python')

plt.legend()
plt.show()
```

```python
# we can change these lines colors also
# using use_line_collection=False
# by default is True

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=1,
         label='Python', use_line_collection=False)

plt.legend()
plt.show()
```

```python
# we can change its orientation also
# using orientation='horizontal'
# by default is 'vertical'

plt.stem(x,y,linefmt='-.', markerfmt='g*', bottom=1,
         label='Python', use_line_collection=False,
        orientation='horizontal')

plt.legend()
plt.show()
```
