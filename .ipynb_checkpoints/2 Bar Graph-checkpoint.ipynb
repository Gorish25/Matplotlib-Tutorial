```python
import matplotlib.pyplot as plt
```

#### creating a linear Graph

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show()
```

#### creating a bar Graph

```python
plt.bar(x,y)
plt.show()
```

```python
# giving color
c=['r','y','g','b']
plt.bar(x,y,color=c)
plt.show()
```

```python
# creating another bar plot
x=['Python','c','c++','Java']
y=[85,70,60,82]

# colors
c=['r','y','g','b']

# labels
# with fontsize
plt.xlabel("Languages", fontsize=20)
plt.ylabel('Numbers', fontsize=20)

# title
plt.title("Online Data for Programming Languages", fontsize=30)
print()

# plotting bar graph
plt.bar(x,y,color=c)
plt.show()
```

```python
# changing width of bars
plt.bar(x,y,width=0.2)
plt.show()
```

```python
# converting colors
plt.bar(x,y,color='g')
plt.show()
```

```python
# converting all bars colors
c=['r','y','m','b']
plt.bar(x,y,color=c,width=0.5)
plt.show()
```

```python
# shifting x axis bar names to bar edges
# by default is align='center'
plt.bar(x,y,align='edge', width=0.5, color=c)
plt.show()
```

```python
# changing edge color
plt.bar(x,y,color=c,edgecolor='b')
plt.show()
```

```python
# changing edge width
plt.bar(x,y,color=c, edgecolor='black', linewidth=5)
```

```python
# changing edge style
# for dotted we can use : also
plt.bar(x,y, color=c, edgecolor='black', linewidth=5, linestyle='dashed')
plt.show()
```

```python
# color are too bright we can convert them to light using
# alpha parameter
plt.bar(x,y, color=c, edgecolor='black', linewidth=5, linestyle=":", alpha=0.5)
plt.show()
```

```python
# giving bar graph a label by label attribute
# we must use plt.legend() when we are using label attribute
plt.bar(x,y, color=c, edgecolor='black', label='Bar Graph of Popularity')
plt.legend()
plt.show()
```

```python
# creating multiple bar Graph in a Graph
w=[20,30,40,35]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Numbers", fontsize=15)
plt.bar(x,y,color=c, label="Popularity", width=0.4)
plt.bar(x,w, width=0.4, color='g', label='Popularity')
plt.legend()
plt.show()
```

```python
# Creating multiple Bar Graph in a Graph without overlapping
import numpy as np

x = ['Python','c','c++','Java']
y = [85,70,60,82]
z = [20,30,40,35]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)

plt.bar(p,y, width, color=c, label='Popularity 1', edgecolor='black')
plt.bar(p1,z, width, color='g', label='Popularity 2', edgecolor='black')

plt.legend()
plt.show()
```

```python
# in above example names of languages are not shown 
# so in below example added some code to show languages names 
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)

plt.bar(p,y, width, color=c, label='Popularity 1', edgecolor='black')
plt.bar(p1,z, width, color='g', label='Popularity 2', edgecolor='black')

plt.xticks(p+width/2,x)

plt.legend()
plt.show()
```

```python
# rotating the names of languages
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)

plt.bar(p,y, width, color=c, label='Popularity 1', edgecolor='black')
plt.bar(p1,z, width, color='g', label='Popularity 2', edgecolor='black')

plt.xticks(p+width/2, x, rotation=20)

plt.legend()
plt.show()
```

```python
# horizontal bar graph
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)

plt.barh(x,y, width, color='r', label='Popularity 1', edgecolor='black')
plt.bar(x,z, width, color='g', label='Popularity 2', edgecolor='black')

plt.legend()
plt.show()
```
