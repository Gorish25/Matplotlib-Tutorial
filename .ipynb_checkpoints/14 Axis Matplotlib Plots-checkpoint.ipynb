```python
import matplotlib.pyplot as plt
```

# Axis Matplotlib Plots
- xticks(), yticks(), xlim(), ylim(), axis()

```python
# creating a simple Graph
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.show()
```

```python
# making some changes in x axis

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# changing x axis value types from float to int
plt.xticks(x)

plt.show()
```

```python
# creating changes to y axis

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# changing y axis value types from float to int
plt.yticks(x) # use x here

plt.show()
```

```python
# giving change to both axis

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# making change to both
plt.xticks(x)
plt.yticks(x) # use x here

plt.show()
```

```python
# showing labels on x axis

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# making change to both
plt.xticks(x, labels=['Python','c','c++','java','r'])
plt.yticks(x) # use x here

plt.show()
```

```python
# increasing x axis limit

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# increasing x axis limit
plt.xlim(0,10)

plt.show()
```

```python
# increasing limit to both axis

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

# increasing x axis limit
plt.xlim(0,10)

# increasing y axis limit
plt.ylim(0,10)

plt.show()
```

```python
# increasing both limits using
# plt.axis([0,x,0,y])

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.axis([0,8,0,10])

plt.show()
```
