```python
from matplotlib import pyplot as plt
```

## Histogram Chart

```python
# creating data for Histogram chart
import numpy as np
import random

x=np.random.randint(10,60,50) # from 10 to 60 we need 50 data
print(x)
```

```python
no=[47,32,55,41,25,28,7,12,57,33,16,42,18,21,25,22,24,12,57,54,35,38,11,51
    ,16,24,21,40,39,49,15,25,26,51,51,23,11,46,34,39,46,27,49,57,49,58,57,44
,43,16]

# plotting histogram
plt.hist(no)
plt.show()
```

```python
# giving title and labels
plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no)
plt.show()
```

```python
# changing color of histogram
plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, color='g')
plt.show()
```

```python
# above diagram have bars but we can't see them 
# for showing the bars we can use edgecolor='value'
plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, color='g', edgecolor='r')
plt.show()
```

```python
# in above example the 10,20 ... 60 are coming but not fixed
# to fix them first we create a list with values 10 to 60
# then assign it to plt.hist(bins=list)

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, color='g', bins=l, edgecolor='r')
plt.show()
```

```python
# giving this data an upper range (0,200) (means range = 0 to 200)
# here we are using 'auto'(means bins='auto') instead of bins=l
# 

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, 'auto', (0,200), edgecolor='r', color='g')
plt.show()
```

```python
# for arranging the above histogram into descending order 
# we use cumulative=-1

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', cumulative=-1)
plt.show()
```

```python
# in the above histograms y axis is starting from 0
# we can change this using bottom=value

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', bottom=10)
plt.show()
```

```python
# we can also give align parameter for changing 10,20 ... 60
# values position on x axis
# align='value' (left, mid, right)

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', align='left')
plt.show()
```

```python
# changing the type of above histogram
# using histtype='value' (step,bar, stepfilled, barstacked)

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', histtype='step')
plt.show()
```

```python
# in above examples orientation is vertical
# we can convert it into horizontal also
# using orientation='value' (horizontal, vertical)

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', orientation='horizontal')
plt.show()
```

```python
# we can also change it's width using
# rwidth=float (between 0 to 1)

l=[10,20,30,40,50,60]

plt.title("Random Data")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.hist(no, edgecolor='r', color='g', rwidth=0.8)
plt.show()
```

```python
# increasing the sizes of xlabel, ylabel and title

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g')
plt.show()
```

```python
# converting it into a log table

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g', log=True)
plt.show()
```

```python
# giving it a label
# if we are using label we must use plt.legend()

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g', label="Python")

plt.legend()
plt.show()
```

```python
# dividing the whole histogram using 
# plt.axvline(location, color)

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g', label="Python")

plt.axvline(35, color='b')

plt.legend()
plt.show()
```

```python
# we can also give axvline a label

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g', label="Python")

plt.axvline(35, color='b', label='line')

plt.legend()
plt.show()
```

```python
# we can also give them a grid

l=[10,20,30,40,50,60]

plt.title("Random Data", fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.hist(no, edgecolor='r', color='g', label="Python")

plt.axvline(35, color='b', label='line')

plt.grid()

plt.legend()
plt.show()
```
