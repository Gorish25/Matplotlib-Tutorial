```python
import matplotlib.pyplot as  plt
```

## Bar Plot

```python
x = ['Python','c','c++','Java']
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
# giving label and title
plt.xlabel('Languages')
plt.ylabel('Numbers')
plt.title('Popularity Bar Graph')
plt.bar(x,y)
plt.show()
```

```python
# changing font size of labels and title
plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)
plt.bar(x,y)
plt.show()
```

```python
# changing width of bars
plt.xlabel('Languages', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title('Popularity Bar Graph', fontsize=20)
plt.bar(x,y, width=0.4)
plt.show()
```

```python
# changing the color of the bars
plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.title("Popularity of Languages", fontsize=20)
plt.bar(x,y, width=0.4, color='r')
plt.show()
```

```python
# giving multiple colors to bars
c=['r','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c)
plt.show()
```

```python
# moving ground dashes of bars to edges
# by default is align='center'
c=['r','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, align='edge')
plt.show()
```

```python
# changing bars edge boundry color
c=['r','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, edgecolor='r')
plt.show()
```

```python
# in above example edge is thin 
# giving that some more width
c=['y','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, edgecolor='r', linewidth=10)
plt.show()
```

```python
# in above bar edges lines are solid 
# we can give it dashed, or dotted
# linestyle="value"(dotted, dashed etc)
c=['m','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, edgecolor='r', linewidth=5, linestyle='dashed')
plt.show()
```

```python
# in above the colors are too bright 
# we can give it less bright by using
# alpha = value (value should be between 0 and 1)
c=['m','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, edgecolor='r', linewidth=5, linestyle='dashed', alpha=0.5)
plt.show()
```

```python
# giving a lebal to above diagram
# we use label='value' attribute
# Note that when we are using label
# we must use plt.legend() also
c=['m','g','b','brown']

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color=c, edgecolor='r', label='Languages description')
plt.legend()
plt.show()
```

```python
# In 1 diagram giving multiple bar graph
z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(x,y, width=0.4, color='r', label='Popularity 1')
plt.bar(x,z, width=0.4, color='g', label='Popularity 2')

plt.legend()
plt.show()
```

```python
# these bar graph are overlapping on one another 
# we want them seperate and touched to one another
# In above examples variable x contains the languages
# this variable x contains a list of languages 
# which also have index 0,1,2,3
# in given example we are creating a list of index 
# by using numpy and containing it in p variable
import numpy as np

p = np.arange(len(x))
# now in plt.bar() replace x with p


z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(p,y, width=0.2, color='r', label='Popularity 1')
plt.bar(p,z, width=0.2, color='g', label='Popularity 2')

plt.legend()
plt.show()
```

```python
# doesnot make sense
# In above example x axis now has values instead of Python,c,c++,java
# and bars are not seperated also

# now we create another variable p1 and add some more width
# so that bars are now seperated
width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]
# now replace 2nd plt.bar(p) by plt.bar(p1)


z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(p,y, width=0.2, color='r', label='Popularity 1')
plt.bar(p1,z, width=0.2, color='g', label='Popularity 2')

plt.legend()
plt.show()
```

```python
# still not solved because we do not have python,c,c++,java
# for that we use plt.xticks(p+width/2,x)

width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]

z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(p,y, width=0.2, color='r', label='Popularity 1')
plt.bar(p1,z, width=0.2, color='g', label='Popularity 2')

plt.xticks(p+width/2,x)

plt.legend()
plt.show()
```

```python
# we can also rotate the names
# by using plt.xticks(p+width/2,x,rotation=integer)

width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]

z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.bar(p,y, width=0.2, color='r', label='Popularity 1')
plt.bar(p1,z, width=0.2, color='g', label='Popularity 2')

plt.xticks(p+width/2,x, rotation=25)

plt.legend()
plt.show()
```

```python
# creating horizontal bar graphs
# by using plt.barh()
# first remove width parameter

width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]

z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.barh(p,y, color='r', label='Popularity 1')
plt.bar(p1,z, color='g', label='Popularity 2')

plt.xticks(p+width/2,x, rotation=25)

plt.legend()
plt.show()
```

```python
# using plt.barh() for both

width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]

z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.barh(p,y, width, color='r', label='Popularity 1')
plt.barh(p1,z, width, color='g', label='Popularity 2')

plt.xticks(p+width/2,x, rotation=25)

plt.legend()
plt.show()
```

```python
# by removing plt.xticks()

width=0.2
p = np.arange(len(x))
p1 =[i+width for i in p]

z=[30,50,60,40]

plt.xlabel("Languages", fontsize=15)
plt.ylabel('Numbers', fontsize=15)

plt.title("Popularity of Languages", fontsize=20)

plt.barh(p,y, width, color='r', label='Popularity 1')
plt.barh(p1,z, width, color='g', label='Popularity 2')

plt.legend()
plt.show()
```
