```python
import matplotlib.pyplot as plt
```

# Pie Chart

```python
# creating Pie Chart using some values

# both must be same
x=[10,20,30,40]
y=['c','c++','java','python']

plt.pie(x)

plt.show()
```

```python
# giving above diagram labels

x=[10,20,30,40]
y=['C','C++','Java','Python']

plt.pie(x, labels=y)

plt.show()
```

```python
# deattaching the above parts of Pie chart
# using explode=list of values

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0.4,0,1.6]  

plt.pie(x, labels=y, explode=ex)

plt.show()
```

```python
# exploding only 1

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0,0,0]  

plt.pie(x, labels=y, explode=ex)

plt.show()
```

```python
# exploding at equal length

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0.2,0.2,0.2]  

plt.pie(x, labels=y, explode=ex)

plt.show()
```

```python
# we can also change their colors

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0,0,0]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c)

plt.show()
```

```python
# we can also show their percentages
# using autopct ='%0.1f%%'

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0,0,0]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%')

plt.show()
```

```python
# we can also give them shadow
# using shadow = True
# but the shadow form will be less

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0,0,0]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True)

plt.show()
```

```python
# we can also increase or decrease the size of above circle

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.2,0,0,0]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, radius=1.5)

plt.show()
```

```python
# seperating all

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, radius=1.5)

plt.show()
```

```python
# in above labels are too close 
# we can give them a distance also
# using labeldistance = float
# using this they will be close or far

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, labeldistance=1.5)

plt.show()
```

```python
# we can give label inside the pie chart also
# using same (labeldistance=float)

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, labeldistance=0.8, radius=1.5)

plt.show()
```

```python
# we can rotate above pie chart also 
# by using startangle=angle (0 to 360)
# default is 0

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, labeldistance=0.8, radius=1.5, startangle=90)

plt.show()
```

```python
# in above text sizes of c,c++ etc are less 
# we can increase or decrease them
# using (textprops={'fontsize':value})
# it receives value as dictionary

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.7, radius=2.5, startangle=90,
       textprops={'fontsize':15})

plt.show()
```

```python
# above pie chart is in anti-clockwise direction
# we can give this a clockwise direction also
# using counterclock= (True or False)
# by default is True

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=1.5, startangle=90,
       textprops={'fontsize':15}, counterclock=False)

plt.show()
```

```python
# we can also increase it shadow and increase its linewidth
# using wedgeprops = {'linewidth':integer}
# it receives value as a dictionary
# we can also give {'width':integer}
# inside the value

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=1.5, startangle=90,
       textprops={'fontsize':15}, counterclock=False, 
        wedgeprops={'linewidth':5})

plt.show()
```

```python
# we can also give edges a color
# inside same wedgeprops

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=1.5, startangle=90,
       textprops={'fontsize':15}, counterclock=False, 
        wedgeprops={'linewidth':5,'edgecolor':'y'})

plt.show()
```

```python
# we can also enhance center
# using center=(a,b)

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=1.5, startangle=90,
       textprops={'fontsize':15}, counterclock=False, 
        wedgeprops={'linewidth':5,'edgecolor':'y'},
       center=(9,7))

plt.show()
```

```python
# we can also rotate labels
# using rotatelabels=True
# by default is False

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=1.5, startangle=90,
       textprops={'fontsize':15}, counterclock=False, 
        wedgeprops={'linewidth':5,'edgecolor':'y'},
       rotatelabels=True)

plt.show()
```

```python
# giving above diagram a title and a label
# must use plt.legend() when using label
# with changing location plt.legend(loc=integer) (between 1 to 10)

x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.1,0.1,0.1,0.1]  

c=['r','g','m','b']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%',
        shadow=True, labeldistance=0.9, radius=0.8, startangle=90,
       textprops={'fontsize':15}, counterclock=False, 
        wedgeprops={'linewidth':5,'edgecolor':'y'},
       rotatelabels=True,)

plt.title('Popularity Chart')
plt.legend(loc=4)

plt.show()
```

```python
# a simple dot pie chart
plt.pie([1])
plt.show()
```

```python
# donnut pie chart or multiple pie chart

x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','python']

plt.pie(x, labels=y, radius=1.5)
plt.pie(x1, radius=0.5)

plt.show()
```

```python
# giving a nice color

x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','python']
c=['r','b','g','y']

plt.pie(x, labels=y, radius=1.5)
plt.pie(x1, radius=1, colors=c)

plt.show()
```

```python
# a complete white ring inside pie chart

x=[10,20,30,40]
y=['c','c++','java','python']

plt.pie(x, labels=y, radius=1.5)
plt.pie([1], colors='w')

plt.show()
```
