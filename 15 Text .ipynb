```python
import matplotlib.pyplot as plt
```

# Working with Text

```python
from IPython import display
display.Image('17img.png')
```

```python
# creating a linear Plot
# with specifications

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

plt.show()
```

```python
# giving corner points a text using
# plt.text(x_position, y_position, text)

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# giving text
plt.text(2.5,5,'java', fontsize=12)
plt.text(4.5,6,'Python', fontsize=12,style='italic')

plt.show()
```

```python
# enclosing text in a box
# using bbox={'facecolor':'color'}

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# giving text
plt.text(2.5,5,'java', fontsize=12, style='italic', 
        bbox={'facecolor':'red'})
plt.text(4.5,6,'Python', fontsize=12,style='italic',
        bbox={'facecolor':'red'})

plt.show()
```

```python
# pointing an arrow to any point
# using annotate
# plt.annotate('text',xy=(pointing position), xytext=(textposition),
# arrowprops=dict(facecolor='color', shrink=integer))

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# giving text
plt.text(2.5,5,'java', fontsize=12)

# pointing arrow
plt.annotate('Python', xy=(5,6), xytext=(3.5,6),
            arrowprops={'facecolor':'black','shrink':100})

plt.show()
```

```python
# using legend

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'])

plt.show()
```

```python
# changing the location of 'up' using loc

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'], loc=5)

plt.show()
```

```python
# changing facecolor of 'up' using
# facecolor = 'color'

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'], loc=2, facecolor='red')

plt.show()
```

```python
# giving 'up' an edge color
# using edgecolor = 'color'


x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'], loc=2, facecolor='red',edgecolor='green')

plt.show()
```

```python
# reducing brightness using 
# framealpha = float

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'], loc=2, facecolor='red',edgecolor='green',
          framealpha=0.5)

plt.show()
```

```python
# Giving 'up' a 3D look using 
# shadow = True

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

# title
plt.title('Gverm',fontsize=20)

# labels of x axis and y axis
plt.xlabel('Days',fontsize=15)
plt.ylabel('Python', fontsize=15)

# using legend
plt.legend(['up'], loc=2, facecolor='red',edgecolor='green',
          framealpha=0.5, shadow=True)

plt.show()
```
