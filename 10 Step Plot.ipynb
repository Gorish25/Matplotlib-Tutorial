```python
import matplotlib.pyplot as plt
```

# Step Plot

```python
# creating value for step plot and plotting Step Plot

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
# changing some values and creating new step plot

x = [4,5,8,1,9,2]
y = [1,8,3,7,6,2]

plt.step(x,y)
plt.show()
```

```python
# giving graph a grid with changing values

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)
plt.show()
```

```python
# Giving graph a title

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)
plt.title('Step Plot')
plt.show()
```

```python
# Giving graph x axis and y axis

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)
plt.title('Step Plot')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# changing font size of labels and title

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.show()
```

```python
# applying marker for points
# using marker='value'

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,marker='*')

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.show()
```

```python
# changing colors

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color='r',marker='*')

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.show()
```

```python
# changing marker size
# using ms=integer

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color='r',marker='*',ms=10)

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.show()
```

```python
# changing marker face color
# using mfc='color'

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color='r',marker='o',ms=10, mfc='g')

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.show()
```

```python
# Giving a label
# must use legend

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color='r',marker='o',ms=10, mfc='g',label='Python')

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.legend()

plt.show()
```

```python
# changing the location of label
# using plt.legend(loc=integer)

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color='r',marker='o',ms=10, mfc='g',label='Python')

plt.title('Step Plot', fontsize=20)

plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.legend(loc=6)

plt.show()
```
