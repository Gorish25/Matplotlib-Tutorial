```python
import matplotlib.pyplot as plt
```

# Box Plot

```python
# creating data for box plot
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
# giving a notch to above diagram
# using notch=True
# by default is False

x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)
plt.show()
```

```python
# above diagrams are vertical boxplots
# it can be horizontal using
# vert=False
# by default is True

x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert=False)
plt.show()
```

```python
# we can increase its width also
# using widths=float value

x = [10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.8)
plt.show()
```

```python
# there is a 1 coming in above example
# we can change it by using labels=['']
# remember that we use list for values
# because boxplot plots multiple list of values

x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'])
plt.show()
```

```python
# In above examples box plot do not having color
# we can fill a color in this by using
# patch_artist=True
# by default is False

x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'], patch_artist=True)
plt.show()
```

```python
# seeing mean using
# showmeans = True
# by default is False

x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'], showmeans=True)
plt.show()
```

```python
# showing outliers by adding a big value to x

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True)
plt.show()
```

```python
# when we add 2 or more than 2 big values in x
# then this outliers will not be shown

x = [10,20,30,40,50,60,70,120,130]

plt.boxplot(x, labels=['Python'], showmeans=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True)
plt.show()
```

```python
# the whisk part is not connected to the outlier
# we can connect it by using
# whis = float

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, whis=1.7)
plt.show()
```

```python
x = [10,20,30,40,50,60,70, 120]

plt.boxplot(x, labels=['Python'], showmeans=True)
plt.show()
```

```python
# we can change the outlier color and diagram also by using
# sym='value' (value contains color and type)

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*')
plt.show()
```

```python
# we can hide the outlier also
# using sym=' '

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym=' ')
plt.show()
```

```python
# we can change the box edge color also using
# boxpros={}

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*',
           boxprops=dict(color='r'))
plt.show()
```

```python
# we can also change the whisk cap color
# using capprops={}

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*',
           boxprops=dict(color='r'), capprops=dict(color='g'))
plt.show()
```

```python
# changing the whisk edge color
# using whiskerprops={}

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*',
           boxprops=dict(color='r'), capprops=dict(color='g'),
           whiskerprops=dict(color='y'))
plt.show()
```

```python
# flierprops=dict(markeredgecolor='value')

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*',
           boxprops=dict(color='r'), capprops=dict(color='g'),
           whiskerprops=dict(color='y'), 
           flierprops=dict(markeredgecolor='brown'))
plt.show()
```

```python
# multiple box plot on same box plot

x = [10,20,30,40,50,60,70,120]
y=[10,20,30,40,50,60,90]


plt.boxplot(x, labels=['Python'], showmeans=True, sym='g*',
           boxprops=dict(color='r'), capprops=dict(color='g'),
           whiskerprops=dict(color='y'))

plt.boxplot(y, labels=['Python'])


plt.show()
```

```python
# boxplot right way

x = [10,20,30,40,50,60,70,120]
y=[10,20,30,40,50,60,90]

z=[x,y]

plt.boxplot(z, labels=['python','c++'], showmeans=True, sym='g*',
           boxprops=dict(color='r'), capprops=dict(color='g'),
           whiskerprops=dict(color='y'))



plt.show()
```
