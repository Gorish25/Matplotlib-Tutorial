```python
from matplotlib import pyplot as plt
```

# Savefig
- it is used to save any plot in computer memory
- plt.savefig(fname)

```python
# using savefig to save any plot in memory

# creating values for graph

x = [1,2,3,4,5]
y = [2,3,6,8,1]

# linear plot
plt.plot(x,y, color='g')

# saving figure
# dpi = integer (for larging the graph and enhancing quality)
# dpi(dot per inch)
# facecolor = 'color' (giving boundry a color)
plt.savefig('13 Linear Plot', dpi=2000, facecolor='red')

plt.show()
```

```python
# saving file as pdf
# we can change it in any format ex .jpg, .pdf, .png
# by default is .png

x = [1,2,3,4,5]
y = [2,3,6,8,1]

# linear plot
plt.plot(x,y, color='g')

# saving figure
# dpi = integer (for larging the graph and enhancing quality)
# dpi(dot per inch)
# facecolor = 'color' (giving boundry a color)
plt.savefig('14 Linear Plot.pdf', dpi=2000, facecolor='red')

plt.show()
```

```python
# creating a transparent file
# using transparent = True
# by default it False

x = [1,2,3,4,5]
y = [2,3,6,8,1]

# linear plot
plt.plot(x,y, color='g')

# saving figure
plt.savefig('15 Linear Plot', dpi=2000, facecolor='red',
           transparent=True)

plt.show()
```

```python
# changing box inches
# using bbox_inches = 'tight'

x = [1,2,3,4,5]
y = [2,3,6,8,1]

# linear plot
plt.plot(x,y, color='g')

# saving figure
plt.savefig('16 Linear Plot', dpi=2000, facecolor='red',
           transparent=False, bbox_inches='tight')

plt.show()
```
