```python
import matplotlib.pyplot as plt
```

# Fill Between Plot
- first use plt.plot(x,y) ( we can use any plot)
- then use plt.fill_between(x,y)

```python
# creating values for fill between plot and plotting also

# without fill between plot example

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# with fill between plot example

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# plotting a linear plot
plt.plot(x,y,color='g')

# fill between plot
plt.fill_between(x,y)


# title
plt.title('Fill Between Plot')

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# giving another values to fill between

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# fill between plot
plt.fill_between(x,y1=2,y2=4)

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# giving it a different color

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# fill between plot
plt.fill_between(x,y1=2,y2=4, color='r')

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# using where keyword
# remember that it do not work with list
# it works well with numpy array
# below example will give you an error
# because it uses list

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# fill between plot
plt.fill_between(x,y1=2,y2=4, color='r', where=(x>=2) & (x<=4))

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# using numpy array error can be resolved

import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# fill between plot
plt.fill_between(x,y1=2,y2=4, color='r', where=(x>=2) & (x<=4))

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# decreasing brightness using alpha=float

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

# plotting a linear plot
plt.plot(x,y,color='g')


# title
plt.title('Fill Between Plot')

# fill between plot
plt.fill_between(x,y1=2,y2=4, color='r', where=(x>=2) & (x<=4),
                alpha=0.5)

# labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
