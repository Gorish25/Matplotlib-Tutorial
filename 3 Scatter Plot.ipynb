```python
import matplotlib.pyplot as plt
```

 ## Scatter Plot

```python
day = [1,2,3,4,5,6,7]
number = [2,3,1,4,5,3,6]
plt.scatter(day,number)
plt.show()
```

```python
# giving title and label to scatter plot to scatter plot
plt.scatter(day,number)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing the color of dots
plt.scatter(day,number, color='r')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# giving different color to every dot
colors=['r','y','g','b','r','g','y']
plt.scatter(day,number, c=colors)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing every dot size to different one
colors=['r','y','g','b','r','g','y']
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing transparancy
# alpha should be between 0 to 1
colors=['r','y','g','b','r','g','y']
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, alpha=0.5)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing dot shape
colors=['r','y','g','b','r','g','y']
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, marker='*')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing dots edge color
colors=['r','y','g','b','r','g','y']
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, edgecolor='g')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# changing edge width
colors=['r','y','g','b','r','g','y']
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, edgecolor='g', linewidth=5)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# we can give dot colors using values also between 0 to 100
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, cmap='viridis')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.show()
```

```python
# with specifying color bar
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, cmap='viridis')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.colorbar()
plt.show()
```

```python
# another example by changing cmap value
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, cmap='BrBG')
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.colorbar()
plt.show()
```

```python
# changing transparancy
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]
plt.scatter(day,number, c=colors, s=sizes, cmap='BrBG', alpha=0.5)
plt.title('Scatter Plot', fontsize=20)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)
plt.colorbar()
plt.show()
```

```python
# giving colorbar a name
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]

plt.scatter(day,number, c=colors, s=sizes, cmap='BrBG')
plt.title('Scatter Plot', fontsize=20)

plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)

t=plt.colorbar()
t.set_label('Color Bar')

plt.show()
```

```python
# double scatter plot in one plot
colors=[10,49,56,30,25,2,17]
sizes=[200,150,350,400,120,140,180]

number2=[3,4,5,6,1,2,7]

plt.scatter(day,number, c=colors, s=sizes, cmap='BrBG')
plt.scatter(day,number2, color='r', s=sizes, cmap='BrBG')

plt.title('Scatter Plot', fontsize=20)

plt.xlabel('Days', fontsize=15)
plt.ylabel('Numbers', fontsize=15)

t=plt.colorbar()
t.set_label('Color Bar')

plt.show()
```
