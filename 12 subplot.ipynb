```python
from matplotlib import pyplot as plt
```

# Subplot
- plt.subplot(nrows,ncols,index)

```python
# creating values for linear plot

# subplot must use before plotting every graph

x = [1,2,3,4]
y = [1,2,3,4]

# linear plot
plt.subplot(2,2,1)
plt.plot(x,y,color='g')

# pie plot
plt.subplot(2,2,2)
plt.pie([1],colors='r')

# advanced pie plot
plt.subplot(2,2,3)
x1 = [10,20,30,40]
plt.pie(x1)

# bar chart
plt.subplot(2,2,4)
x2 = ['a','s','d','f']
y2 = [4,2,3,5]
plt.bar(x2,y2)


plt.show()
```
